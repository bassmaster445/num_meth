```python
import numpy as np
import matplotlib.pyplot as plt

# Define the right-hand side of the differential equation
def RHS(t, y):
    return -t * y + y**3 / 20  # Use np.abs to handle negative y

# Create meshgrid
t, y = np.meshgrid(np.linspace(0, 3, num=30), np.linspace(-3, 3, num=30))

# Compute dt and dy
dt = np.ones(t.shape)  # Unit time step
dy = RHS(t, y)         # dy/dt from the differential equation

# Normalize for quiver plot
pp = 0.5
nn = (dt**2 + dy**2)**pp  # Normalization factor
dt_normalized = dt / nn   # Normalize dt
dy_normalized = dy / nn   # Normalize dy

# Create the quiver plot
plt.figure(figsize=(8, 6))
plt.quiver(t, y, dt_normalized, dy_normalized, color='blue', scale=20)
plt.grid(True)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Direction Field for dy/dt = 5 - 2 * sqrt(|y|)')
plt.show()
```
